```
import json
import requests 
from config import api_key
from pprint import pprint
import pandas as pd
from bs4 import BeautifulSoup
import pymongo
import json
import datetime 
```

```
# Establish connection 
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```
# Define database 
db = client.top_stories_nyt

# Query business articles
business = db.business.find()

# Iterate through the collection
#for articles in business:
    #pprint(articles)
```

```
print(business)
```

```
#lists
title_business = []
published_date_business = []
```

```
# query business articles 
for articles in business:
    print(articles)
    title_business.append(articles["title"])
    published_date_business.append(articles["published_date"])
```

```
# Create business df
business_df = pd.DataFrame(columns = [
    "Title",
    "Published Date",
    ])


business_df["Title"] = title_business
business_df["Published Date"] = published_date_business

business_df
```
