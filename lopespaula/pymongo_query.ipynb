```python
# Module used to connect Python with MongoDb
import pymongo
from pprint import pprint
```

```python
# The default port used by MongoDB is 27017
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the database in Mongo
db = client.most_popular_articles_nyt_db
```

```python
# Query most viewed articles
most_viewed = db.most_viewed.find()

# Iterate through each student in the collection
for articles in most_viewed:
    pprint(articles)
```

```python
# Query most emailed articles
most_emailed = db.most_emailed.find()

# Iterate through each student in the collection
for articles in most_emailed:
    pprint(articles)
```

```python
# Query most shared articles
most_shared = db.most_shared.find()

# Iterate through each student in the collection
for articles in most_shared:
    pprint(articles)
```
