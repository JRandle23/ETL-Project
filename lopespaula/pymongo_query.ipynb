```python
import pandas as pd
import pymongo
from pprint import pprint
```

```python
# The default port used by MongoDB is 27017
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the database in Mongo
db = client.most_popular_articles_nyt_db
```

```python
# Query most viewed articles
most_viewed = db.most_viewed.find()

# Iterate through the collection
# for articles in most_viewed:
#     pprint(articles)
```

```python
print(most_viewed)
```

```python
#lists
title_viewed = []
published_date_viewed = []
section_viewed = []
subsection_viewed = []
keywords_viewed = []
```

```python
for articles in most_viewed:
    print(articles)
    title_viewed.append(articles["title"])
    published_date_viewed.append(articles["published_date"])
    section_viewed.append(articles["section"])
    subsection_viewed.append(articles["subsection"])
    keywords_viewed.append(articles["adx_keywords"])
    
```

```python
viewed_df = pd.DataFrame(columns = [
    "Title",
    "Published Date",
    "Section",
    "Subsection",
    "Keywords"
    ]) 
```

```python
viewed_df["Title"] = title_viewed
viewed_df["Published Date"] = published_date_viewed
viewed_df["Section"] = section_viewed
viewed_df["Subsection"] = subsection_viewed
viewed_df["Keywords"] = keywords_viewed

viewed_df
```

```python
count_per_section = viewed_df.groupby("Section").count()
print(count_per_section.loc[:,"Title"])
```
